```python
import networkx as nx
import csv
import rdflib
import heapq
from itertools import combinations, permutations
```

```python
edges = []
with open('data/datamusee1000.csv', 'r',encoding='utf-8') as csvfile:
    spamreader = csv.reader(csvfile, quotechar='|',quoting=csv.QUOTE_NONE)
    for row in spamreader:
        edges.append(row)
```

```python
G = nx.DiGraph()
for edge in edges[1:]:
    G.add_edge(edge[0],edge[2], label = edge[1])
```

```python
print(nx.info(G))
```

```python
degrees = dict(G.out_degree())
for node, degree in degrees.items():
    G.node[node]['degree'] = degree

topnode = heapq.nlargest(1000, degrees.keys(), key=lambda k: degrees[k])
subG = G.subgraph(topnode)
```

```python
subG = G.subgraph(nodes)
pos = nx.spring_layout(subG, dim = 2)
nx.draw(subG, pos = pos, node_size= 1)
nx.draw_networkx_nodes(subG, pos = pos, nodelist= nodes.difference(topnode))
```

```python
center = nx.betweenness_centrality(G)
topcenter = heapq.nlargest(100, center.keys(), key=lambda k: degrees[k])
subG = G.subgraph(topcenter)
nodes = set()
gr = nx.Graph(G)
for s,v in combinations(topcenter, 2):
    if(nx.has_path(gr,s,v)):
        path = nx.shortest_path(gr, s,v)
        nodes.update(path)
    print(len(nodes))
nodes.update(topcenter)

subG = G.subgraph(nodes)
pos = nx.spring_layout(G, dim = 2)
nx.draw(subG, pos = pos, node_size= 5)
nx.draw_networkx_nodes(subG, pos = pos, nodelist= nodes.difference(topcenter))
print(len(nodes))
```

```python
edges = []
with open('data/sembib.json', 'r',encoding='utf-8') as jsonfile:
    spamreader = json.load(jsonfile)
```

```python
head = spamreader['head']['vars']
edges_data = spamreader['results']['bindings']
G_sembib = nx.MultiDiGraph()
for edge in edges_data:
    v = edge[head[0]]['value']
    label = edge[head[1]]['value']
    s = edge[head[2]]['value']
    G_sembib.add_edge(v,s, label = label, type = edge[head[1]]['type'] )
```

```python
print(nx.info(G_sembib))
```

```python
pos = nx.spring_layout(G_sembib)
nx.draw(G_sembib, pos = pos, node_size=2)
```

```python
degrees = dict(G_sembib.out_degree())
for node, degree in degrees.items():
    G_sembib.node[node]['degree'] = degree

topnode = heapq.nlargest(10, degrees.keys(), key=lambda k: degrees[k])

nodes = set()
gr = nx.Graph(G_sembib)
for s,v in combinations(topnode, 2):
    if(nx.has_path(gr,s,v)):
        path = nx.shortest_path(gr, s,v)
        nodes.update(path)
nodes.update(topnode)

subG = G_sembib.subgraph(nodes)
pos = nx.spring_layout(G_sembib, dim = 2)
nx.draw(subG, pos = pos, node_size= 5)
nx.draw_networkx_nodes(subG, pos = pos, nodelist= nodes.difference(topnode))
print(len(nodes))
```
